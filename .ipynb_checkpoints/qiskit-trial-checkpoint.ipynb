```python
import qiskit as q
from qiskit.tools.visualization import plot_bloch_multivector
from qiskit.visualization import plot_histogram
from matplotlib import style

style.use("dark_background")
%matplotlib inline

qasm_simulator = q.Aer.get_backend("qasm_simulator")
statevector_simulator = q.Aer.get_backend("statevector_simulator")

def do_job(circuit, shots=500):
    result = q.execute(circuit, backend=statevector_simulator).result()
    statevec = result.get_statevector()
    
    circuit.measure([i for i in range(circuit.num_qubits)], [i for i in range(len(circuit.clbits))])
    qasm_result = q.execute(circuit, backend=qasm_simulator, shots=shots).result()
    counts = qasm_result.get_counts()
    
    return statevec, counts
```

```python
circuit = q.QuantumCircuit(2, 2)
statevec, counts = do_job(circuit)

plot_bloch_multivector(statevec)
```

```python
plot_histogram([counts], legend=["output"])
```

```python
circuit = q.QuantumCircuit(2, 2)
circuit.h(1)
circuit.x(1)

statevec, counts = do_job(circuit)
plot_bloch_multivector(statevec)
```

```python
circuit = q.QuantumCircuit(3, 3)
circuit.h(0)
circuit.x(0)
circuit.ccx(0, 1, 2)

statevec, counts = do_job(circuit)
#circuit.draw()
plot_bloch_multivector(statevec)
```

```python
plot_histogram([counts], legend=["output"])
```
